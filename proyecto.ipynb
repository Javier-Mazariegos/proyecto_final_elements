## Breve descripción de los datos

- CUST_ID : Identification of Credit Card holder (Categorical)
- BALANCE : Balance amount left in their account to make purchases (
- BALANCE_FREQUENCY : How frequently the Balance is updated, score between 0 and 1 (1 = frequently updated, 0 = not frequently updated)
- PURCHASES : Amount of purchases made from account
- ONEOFF_PURCHASES : Maximum purchase amount done in one-go
- INSTALLMENTS_PURCHASES : Amount of purchase done in installment
- CASH_ADVANCE : Cash in advance given by the user
- PURCHASES_FREQUENCY : How frequently the Purchases are being made, score between 0 and 1 (1 = frequently purchased, 0 = not frequently purchased)
- ONEOFFPURCHASESFREQUENCY : How frequently Purchases are happening in one-go (1 = frequently purchased, 0 = not frequently purchased)
- PURCHASESINSTALLMENTSFREQUENCY : How frequently purchases in installments are being done (1 = frequently done, 0 = not frequently done)
- CASHADVANCEFREQUENCY : How frequently the cash in advance being paid
- CASHADVANCETRX : Number of Transactions made with "Cash in Advanced"
- PURCHASES_TRX : Numbe of purchase transactions made
- CREDIT_LIMIT : Limit of Credit Card for user
- PAYMENTS : Amount of Payment done by user
- MINIMUM_PAYMENTS : Minimum amount of payments made by user
- PRCFULLPAYMENT : Percent of full payment paid by user
- TENURE : Tenure of credit card service for user

Notas:
- BALANCE: El saldo dispoible para el usuario de la tarjeta.
- BALANCE_FREQUENCY: La frecuencia de uso de la cuenta, esto puede ser gasto o porque adelanto a pagar la tarjeta.
- PURCHASES: El total de todas las compras que se han realizado, sumando compras de un solo pago y compras a plazo. 
- ONEOFF_PURCHASES: El monto maximo que se ha gastado en una sola compra.
    - Esto puede ser cero cuando no se ha realizado un solo pago. 
- INSTALLMENTS_PURCHASES: El monto total de la compra que se realizó a plazos. 
- CASH_ADVANCE: La cantidad total de dinero que se le ha metido a la tarjeta antes de realizar una compra. 
- PURCHASES_FREQUENCY: La fercuencia de compras por parte de los usaurios de las tarjetas de credito.
- ONEOFF_PURCHASES_FREQUENCY: La frecuencia de compras de un solo pago. 
    - Comprobar si hay casos en los que PURCHASES_FREQUENCY sea 0 y ONEOFF_PURCHASES_FREQUENCY tenga algún porcentaje.
    - Comprobar si hay casos en los que ONEOFF_PURCHASES_FREQUENCY sea 0, entonces ONEOFF_PURCHASES debe de ser cero. 
- PURCHASES_INSTALLMENTS_FREQUENCY: La frecuencia en donde las compras se han realizado a plazos.
- CASH_ADVANCE_FREQUENCY: Frecuencia de pagos por adelantado. 
- CASH_ADVANCE_TRX: Numero de transaccion que se han realizado con dinero por adelantado.
    - Funciona como un metodo de pago más seguro, pero se usa más para las exportaciones. Se paga por adelantado, para que el proveedor tenga seguridad de la compra. 
- PURCHASES_TRX: Numero de compras que se han realizado.
- CREDIT_LIMIT: Limite de credito que tiene autorizado para una tarjeta de credito. 
- PAYMENTS: Ultimo pago que realizó el usuario a su tarjeta. 
- MINIMUM_PAYMENTS: Pago minimo que realizó el usuario a su tarjeta.
- PRC_FULL_PAYMENT: Porcentaje del pago que realizó el usuario a su tarjeta contra el total de lo que tenia que pagar.
- TENURE: La cantidad de meses que ha tenido el usuario la tarjeta de credito.

# Análisis de PCA

```python
from sklearn.decomposition import PCA
```

```python
pca_vars = ['BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES',
       'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE',
       'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY',
       'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY',
       'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS',
       'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT', 'TENURE']
```

Verificaremos que la data sea normal antes de hacer el PCA 

```python
import statsmodels.api as sm
from scipy import stats
```

```python
for variable_name in pca_vars: 
    k2, p_value = stats.normaltest(dataset[variable_name])
    print("Variable: ",variable_name)
    print(f"Estadístico = {k2}, p-value = {p_value}")
```

Normalización de variables

```python
X = dataset[pca_vars].values
mu = X.mean(axis=0)
std = X.std(axis=0)
X = (X - mu) / std
```

PCA

```python
pca = PCA()
pca.fit(X)

variances = pca.explained_variance_ratio_
directions = pca.components_

print("Varianzas de cada componente:", variances)
print("Direcciones de cada componente:", directions)
```

```python
varianzas = pd.DataFrame({"variance":variances})
varianzas = varianzas.reset_index().rename(columns={"index":"component"})
varianzas
```

```python
sns.barplot(varianzas, x = "component", y = "variance")
```

```python
pcamodel = PCA()
pca2 = pcamodel.fit_transform(X)
pca2.shape
```

```python
ll = pcamodel.explained_variance_
ll / ll.sum()
```

```python
def mybiplot(score, coeff, labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley,s=5)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'green', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
 
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()
```

```python
plt.figure(figsize=(6,6))
mybiplot(pca2[:,0:2], np.transpose(pcamodel.components_[0:2, :]), list(dataset[pca_vars].columns))
plt.show()
```

```python
plt.bar(range(1,len(pcamodel.explained_variance_ )+1),pcamodel.explained_variance_ )
plt.ylabel('Explained variance')
plt.xlabel('Components')
plt.plot(range(1,len(pcamodel.explained_variance_ )+1),
         np.cumsum(pcamodel.explained_variance_),
         c='red',
         label="Cumulative Explained Variance")
plt.legend(loc='upper left')
```

### Data Split

```python
from sklearn.model_selection import train_test_split

X = dataset[model_vars].drop(['Paga'], axis=1)
y = dataset["Paga"]

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=.30, random_state=123)
```

### Preprocesamiento

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = Pipeline([
                        ("mean_imputer", SimpleImputer(strategy="mean")),
                        ("std_scaler", StandardScaler())
                    ])
```

```python
numeric_vars = X_train.columns

full_transfomer = ColumnTransformer([
        ("numeric", num_pipe, numeric_vars)
    ])
```

```python
X_train_proccesed = full_transfomer.fit_transform(X_train)
X_val_proccesed = full_transfomer.transform(X_val)
```

### Selección de modelo - GridSearch

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score
from sklearn.model_selection import GridSearchCV
```

```python
modelos = [
    AdaBoostClassifier(random_state=21), 
    GradientBoostingClassifier(random_state=21),
    XGBClassifier(random_state= 21),
    RandomForestClassifier(random_state=21, n_jobs=-1)   
]

modelos_name = [
    "AdaBoost",
    "GradientBoosting",
    "XGB",
    "RandomForest"  
]

search_space = [
    {
        "n_estimators" : [100, 200],
        "learning_rate" : [0.1, 0.3, 0.7]
    }, 
    {
        "n_estimators" : [100, 200],
        "learning_rate" : [0.1, 0.3, 0.7], 
        "min_samples_split" : [2, 3, 6], 
        "max_depth": [3, 6, 9]
    }, 
    {
        "n_estimators" : [100, 200],
        "learning_rate" : [0.1, 0.3, 0.7], 
        "gamma" : [0.01, 0.1], 
        "max_depth": [3, 6, 9]
    },
    {
        "n_estimators" : [100, 200],
        "min_samples_split" : [2, 3, 6], 
        "max_depth": [3, 6, 9]
    }
]
```

```python
# Función para encontrar los mejores parámetros para cada modelo usando grid search

def grid_evaluation(modelos, modelos_name, search_space, X_train_p, y_train,i_selector):    
    print("Modelo:", i_selector)

    GS = GridSearchCV(modelos[i_selector], 
                  param_grid=search_space[i_selector], 
                  scoring=["accuracy", "f1"], 
                  refit="f1", 
                  cv=3, 
                  verbose=4, 
                  n_jobs=-1)
    
    GS.fit(X_train_p, y_train)

    print("Mejor modelo:", GS.best_estimator_)
    print("Parametros:  ",GS.best_params_)
    print("Score F1:   ", GS.best_score_, "\n")

    nombre = "resultados_cv"+ modelos_name[i_selector]+".xlsx"
    pd.DataFrame(GS.cv_results_).sort_values("rank_test_f1").to_excel(nombre, index = False)

    return GS
```

```python
grid_evaluation(modelos, modelos_name, search_space, X_train_proccesed, y_train, 0)
```

```python
grid_evaluation(modelos, modelos_name, search_space, X_train_proccesed, y_train, 1)
```

```python
grid_evaluation(modelos, modelos_name, search_space, X_train_proccesed, y_train, 2)
```

```python
grid_evaluation(modelos, modelos_name, search_space, X_train_proccesed, y_train, 3)
```

## Creando categorias de clientes

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

dataset["PURCHASES_MINMAX"] =scaler.fit_transform(dataset[["PURCHASES"]])
```

```python
sns.scatterplot(dataset, x = "PRC_FULL_PAYMENT", y = "PURCHASES_MINMAX")
```

```python
cluster_vars = ["PRC_FULL_PAYMENT", "PURCHASES_MINMAX"]
```

```python
from sklearn.cluster import KMeans

distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(dataset[cluster_vars])
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

El metodo del codo sugiere crear dos grupos para el clustering, pero haremos una prueba con 3 para ver el resultado

```python
from scipy.cluster.hierarchy import dendrogram, linkage

link = linkage(dataset[cluster_vars].sample(frac=0.1, random_state=30), metric='euclidean', method='weighted', optimal_ordering=True)
plt.figure(figsize=(10,8))
D = dendrogram(link, color_threshold=0.4)
plt.show()
```

Pruebas de cual algoritmo categoriza mejor

```python
# Algoritmos de clustering a comparar
from sklearn.cluster import k_means, KMeans, Birch, KMeans, MeanShift
from sklearn.cluster import AgglomerativeClustering, SpectralClustering

km = KMeans(n_clusters=3, random_state=2023)
ms = MeanShift(bandwidth=0.3)
bi = Birch(n_clusters=3)
sp = SpectralClustering(n_clusters=3, affinity='rbf', gamma=1.0)
ah = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')

methods = [km, ms, sp, bi, ah]
```

```python
y_labels = methods[4].fit_predict(dataset[cluster_vars])
sns.scatterplot(dataset, x = "PRC_FULL_PAYMENT", y = "PURCHASES_MINMAX", hue=y_labels)
```
